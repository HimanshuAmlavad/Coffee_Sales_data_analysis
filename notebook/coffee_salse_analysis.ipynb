```python
import pandas as pd
import matplotlib.pyplot as plt

coffee_data = pd.read_csv("C:/Data Analysis Project/Coffee Salse Data Analysis/Data/CoffeeSalseData.csv")
print(coffee_data.head())
```

```python
 coffee_data.info()
```

```python
 coffee_data.isnull().sum()
```

```python
total_sales = coffee_data['money'].sum()
print("Total Sales:", total_sales)
```

```python
transaction_counts = coffee_data['transaction_type'].value_counts()

print(transaction_counts)
```

```python
plt.figure(figsize=(3,3))
plt.pie(transaction_counts, labels=transaction_counts.index, autopct='%1.1f%%', startangle=90)
plt.title("Transaction Type Distribution")
plt.show()
```

```python
most_sold_coffee = coffee_data['coffee_name'].value_counts()
print(most_sold_coffee)
```

```python
coffee_sales = coffee_data.groupby('coffee_name')['money'].sum().sort_values(ascending=False)
print(coffee_sales)
```

```python
# Plot Coffee Sales Count
plt.figure(figsize=(8,4))
most_sold_coffee.plot(kind='bar', color='skyblue')
plt.title("Most Sold Coffee (Count)")
plt.xlabel("Coffee Type")
plt.ylabel("Number of Transactions")
plt.show()

# Plot Coffee Sales by Amount
plt.figure(figsize=(8,4))
coffee_sales.plot(kind='bar', color='orange')
plt.title("Coffee Sales by Revenue")
plt.xlabel("Coffee Type")
plt.ylabel("Total Sales Amount")
plt.show()
```

```python
monthly_sales = coffee_data.groupby('Month')['money'].sum()
print(monthly_sales)
```

```python
plt.figure(figsize=(10,5))
monthly_sales.plot(kind='bar', color='skyblue')
plt.title("Monthly Sales")
plt.xlabel("Month")
plt.ylabel("Total Sales Amount")
plt.xticks(rotation=45)
plt.show()
```
